```python
import pandas as pd
```

```python
data = pd.read_excel("DV_Final Data_without Technologies.xlsx")
```

```python
data.head()
```

```python
survey_data = pd.read_csv('survey_results_public.csv')
maps = pd.read_csv('maps.csv')
survey_data = pd.merge(survey_data, maps,left_on=['Country'], right_on = ['Country'], how = 'left')
survey_data['longitude'].isna().sum()
#print(data[data['longitude'].isnull()])
```

```python
survey_data
```

```python
#merge lat long to data
data = pd.merge(data, survey_data[['Respondent','Country','latitude','longitude']],left_on=['Respondent'], right_on = ['Respondent'], how = 'left')
```

```python
data.head()
```

```python
languages = ['Respondent','LanguageWorkedWith',
             'DatabaseWorkedWith',
             'DatabaseDesireNextYear',
             'PlatformWorkedWith',
             'PlatformDesireNextYear',
             'WebFrameWorkedWith',
             'WebFrameDesireNextYear'
            ]
languages_data = survey_data[languages]
languages_data["skill"] = languages_data["LanguageWorkedWith"] + ";" + languages_data["DatabaseWorkedWith"] + ";" + languages_data["DatabaseDesireNextYear"] + ";" + languages_data["PlatformWorkedWith"] + ";" + languages_data["PlatformDesireNextYear"] + ";" + languages_data["WebFrameWorkedWith"] + ";" + languages_data["WebFrameDesireNextYear"]
```

```python
languages_data = languages_data[['Respondent','skill']]
languages_data.head()
```

```python
%%time
print(languages_data.shape)
languages_data = languages_data.dropna()
print(languages_data.shape)
```

```python
languages_data.head()
```

```python
%%time
skills = languages_data.skill.apply(lambda x: pd.Series(str(x).split(";")))
languages_data = languages_data.merge(skills, left_index=True, right_index=True)
```

```python
data = data.merge(languages_data, on=['Respondent'])
```

```python
data.to_csv("data_with_country_skill.csv")
```

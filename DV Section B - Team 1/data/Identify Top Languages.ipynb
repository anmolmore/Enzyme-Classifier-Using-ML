```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('survey_results_public.csv')
```

```python
languages = ['LanguageWorkedWith',
             'DatabaseWorkedWith',
             'DatabaseDesireNextYear',
             'PlatformWorkedWith',
             'PlatformDesireNextYear',
             'WebFrameWorkedWith',
             'WebFrameDesireNextYear'
            ]
languages_data = data[languages]
languages_data.head()
```

```python
languages_data["language"] = languages_data["LanguageWorkedWith"] + ";" + languages_data["DatabaseWorkedWith"] + ";" + languages_data["DatabaseDesireNextYear"] + ";" + languages_data["PlatformWorkedWith"] + ";" + languages_data["PlatformDesireNextYear"] + ";" + languages_data["WebFrameWorkedWith"] + ";" + languages_data["WebFrameDesireNextYear"]
```

```python
languages_data.head(30)
```

```python
%%time
data_count = languages_data.language.apply(lambda x: pd.Series(str(x).split(";")))
print(data_count)
```

```python
total = []
for i in range(0,115) :
    items = data_count.ix[:, i].tolist()
    total.append(items)

unique_elements, counts_elements = np.unique(total, return_counts=True)
print(np.asarray((unique_elements, counts_elements)))
# unique_words = set(total)
# unique_word_count = len(unique_words)
```

```python
u, count = np.unique(total, return_counts=True)
count_sort_ind = np.argsort(-count)
print(u[count_sort_ind])
print(count[count_sort_ind])
```
